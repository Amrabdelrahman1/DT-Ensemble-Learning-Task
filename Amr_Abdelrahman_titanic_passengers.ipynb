```python
data['Age']=data['Age'].fillna(data['Age'].median())
```

```python
plt.figure(figsize=(15,10))
for col in numeric_col:
  plt.subplot(4,4,numeric_col.index(col)+1)
  sns.kdeplot(data[col])
  plt.tight_layout()
  plt.xlabel(col)
  plt.ylabel('Frequency')
  plt.title(col)
plt.show()
```

```python
for col in data.columns:
  if data[col].dtype==object:
    data[col]=data[col].fillna(data[col].mode()[0])
```

```python
data.duplicated().sum()
```

```python
data.isna().sum()
```

```python
out_col=['Age','Fare']
```

```python
Q1 = data[out_col].quantile(0.25)
Q3 = data[out_col].quantile(0.75)
IQR = Q3 - Q1
outliers = ((data[out_col] < (Q1 - 1.5 * IQR)) |(data[out_col] > (Q3 + 1.5 * IQR))).any(axis=1)
```

```python
outliers.sum()
```

```python
for col in out_col:
  data[col]=np.where(outliers,data[col].mean(),data[col])
```

```python
X = data.drop(['Survived', 'PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
y = data['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
```

```python
for col in X_train.select_dtypes(include='object'):
    X_train[col] = encoder.fit_transform(X_train[col])
    X_test[col] = encoder.transform(X_test[col])
```

```python
y_train = encoder.fit_transform(y_train)

y_test = encoder.transform(y_test)
```
